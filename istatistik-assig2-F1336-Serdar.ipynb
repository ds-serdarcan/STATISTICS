```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sd = 3
n = 36
xbar = 68
alpha = 0.90
```

```python
sem = sd/np.sqrt(n)
sem
```

```python
stats.norm.interval(alpha=alpha,loc=xbar, scale=sem)
```
